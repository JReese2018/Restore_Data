## Skin Service Credit Balance

```python
## Let's start at the beginning and import the libraries and reading the data
import pandas as pd
df_skin = pd.read_csv('balance_skin_services_data.csv')
```

```python
## As always lets look at the data in order to make sure everything is ok
df_skin.head(15)
```

```python
## Now we can get rid of columns that we don't need and redefine the variable
df_skin = df_skin.drop(columns=['Studio Name', 'Day of Issue Date', 'Day of Expire Date', 'Client ID', 'Business Category', 'Credit Source', 'Initial Credit Count', 'Used Credit Count', 'Index', 'Max date'])

## Printing
df_skin.head(10)
```

```python
## We'll also rename the columns to make them easily usable and redefine the varible
df_skin = df_skin.rename(columns={'Full Name': 'full_name'})
df_skin = df_skin.rename(columns={'Phone #': 'phone'})
df_skin = df_skin.rename(columns={'Credit Description': 'credit_description'})
df_skin = df_skin.rename(columns={'Remaining Credits': 'remaining_credits'})

## Printing
df_skin.head(10)
```

```python
## There are going to be some decrepencies with this because some of the data from the 'Remainind Credits' column might not have numbers
## Because of that, I'm going to go ahead and get rid of them as they will mess up the calculations
df_skin = df_skin.dropna()

## Printing
df_skin.head(10)
```

<font size="5">Working with bad data</font>

```python
## We need to get the initial credits in an integer in order to perform calculations
## Since we are wokring with data that does not give us the init
df_skin['initial_credits'] = df_skin['credit_description'].astype(str).str.extract(r'(\d+)', expand=False)

## Printing
df_skin.head(10)
```

```python
## If we tried to iterrate through the rows and do our calculations, it wont work
## Let's see why it won't
df_skin.dtypes
```

```python
## As you can see, the new initial credits is seen as an object
## We can change this by turning it into a string type and then an integer
df_skin['initial_credits'] = df_skin['initial_credits'].astype(str).astype(int)

## Now let's check
df_skin.dtypes
```

```python
## Let start performing calculations
skin_list = []

for index, row in df_skin.iterrows():
    remaining = row['remaining_credits']
    initial = row['initial_credits']
    calculated_percentage = (remaining / initial) * 100
    utilization = 100 - calculated_percentage
    if utilization > 60:
        my_list = [row.full_name, row.phone, row.credit_description, row.remaining_credits]
        skin_list.append(my_list)

## Printing
skin_list
```

## Making the List actionable

```python
## Now that we have the list, let's turn it into a dataframe
core_list_df = pd.DataFrame(skin_list, columns=['Full Name', 'Phone Number', 'Credit Description', 'Remaining Credits'])

## Printing
core_list_df
```

```python
## Since that looks good, we can turn that into an excel sheet and send it to the reps
core_list_df.to_excel('Remaining Skin Credits.xlsx')
```
