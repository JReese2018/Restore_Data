```python
## Load Libs
import pandas as pd
```

```python
## Import Datasets
df = pd.read_csv('feb-march.csv')
df
```

```python
df = df[['First Name', 'Last Name', 'Email', 'Phone #', 'Item', 'Gross adjusted Revenue', 'Credit Used', 'Discount']]
df = df.dropna()
```

```python
df = df.loc[(df['Item'] == 'Base IV Drip') & (df['Credit Used'] == 4) & (df['Discount'] == 132) | (df['Item'] == 'Immunity') | (df['Item'] == 'Energy') | (df['Item'] == 'Recovery') | (df['Item'] == 'Age-Defying')]
df
```

```python
df = df[['First Name', 'Last Name', 'Email', 'Phone #']]
df
```

```python
df = df.reset_index()
df
```

```python
df = df.drop(columns=['index'])
df
```

```python
df.drop_duplicates(subset=['Last Name'], keep='last', inplace=True)
df
```

```python
df.to_excel('Members who have done Base IVs in the past 2 Months.xlsx')
```
